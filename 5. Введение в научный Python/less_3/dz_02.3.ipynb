```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

t = np.arange(0, 2*np.pi, 0.01)          # угол t от 0 до 2pi с шагом 0.01
r = 4                                    # радиус 4

plt.plot(r*np.sin(t), r*np.cos(t), lw=3) 
plt.axis('equal')                        # чтобы круг не был овалом
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
x_left = []
x_right = []
y1 = []
y2 = []
y3 = []
y4 = []
a = 2
b = 1
for i in range(-a*3000, -a*1000):
    x_ = i/1000
    x_left.append(x_)
    y1.append(math.sqrt((x_**2/a**2 - 1)*b**2))
    y2.append(-math.sqrt((x_**2/a**2 - 1)*b**2))
plt.plot(x_left,y1)
plt.plot(x_left,y2)
for i in range(a*1000, a*3000):
    x_ = i/1000
    x_right.append(x_)
    y3.append(math.sqrt((x_**2/a**2 - 1)*b**2))
    y4.append(-math.sqrt((x_**2/a**2 - 1)*b**2)) 
plt.plot(x_right,y3)
plt.plot(x_right,y4)
```
