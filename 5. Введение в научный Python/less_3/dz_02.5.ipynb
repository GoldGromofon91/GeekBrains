```python
%matplotlib inline
import numpy as np
import matplotlib as plt
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)

X, Y = np.meshgrid(X, Y)

Z = 2*Y + 5
Z2 = 2*Y + 10
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
# show()
```
