```python
%matplotlib inline
import math,random
import numpy as np
import matplotlib.pyplot as plt
x=[]
y=[]
r= random.randint(0,100)
for i in range(1,361,1):
    per= math.radians(i)
    x.append(r*np.cos(per))
    y.append(r*np.sin(per))
plt.axis('equal')   
plt.plot(x,y)
```
