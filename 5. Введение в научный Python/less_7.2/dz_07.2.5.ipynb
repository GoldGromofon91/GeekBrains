```python
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
A = np.array([[1,2,-1],[8,-5,2]])
B = np.array([1,12])
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X +2*Y -1
Z2 = 6 - 4*X - 2.5*Y
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
root = np.linalg.lstsq(A,B)
print(root)
```
