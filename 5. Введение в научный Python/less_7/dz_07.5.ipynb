```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
vec_a = np.array([1, 5, 0])
vec_b = np.array([4, 1, 0])
vec_c = np.array([2, 6, 1])
vec_mut = np.cross(vec_a, vec_b)
result = np.inner(vec_mut, vec_c)
print(f'Результат смешанного произведения:\n{result}')
print('-'*60)
print(' '*20 + 'График векторов')
print('-'*60)
X ,Y, Z = np.array([0,0,0]),np.array([0,0,0]),np.array([0,0,0])
U, V, W = np.array([vec_a[0],vec_b[0],vec_c[0]]),np.array([vec_a[1],vec_b[1],vec_c[1]]),np.array([vec_a[2],vec_b[2],vec_c[2]])

fig = plt.figure()
axes = fig.add_subplot(111, projection='3d')

axes.quiver(X,Y,Z,U,V,W)
axes.set_xlim([-10,10])
axes.set_ylim([-10,10])
axes.set_ylim([-1,10])
plt.show()
```
